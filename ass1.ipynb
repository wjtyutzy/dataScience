```python
""" 
Question 1:
Write a Python function named grade_classifier that takes an integer argument (a student's score) and returns the corresponding letter grade.
"""

def grade_classifier(score):
    """
    Grading Scale:
    - Score >= 90: 'A'
    - 80 <= Score < 90: 'B'
    - 70 <= Score < 80: 'C'
    - 60 <= Score < 70: 'D'
    - Score < 60: 'F'
    """
    if score >= 90:
        return 'A'
    elif 80 <= score < 90:
        return 'B'
    elif 70 <= score < 80:
        return 'C'
    elif 60 <= score < 70:
        return 'D'
    else:
        return 'F'


# Test the function with various input scores
test_scores = [100, 99, 82, 79, 70, 61, 60, 59]
for score in test_scores:
    print(f"For a score of {score}, the corresponding grade is: {grade_classifier(score)}")
```

```python
"""
Qustion 4:
Extend the grade_classifier function to include an optional argument that allows you to choose between a standard grading scale and a custom
grading scale defined as a dictionary where the keys are score ranges and the values are corresponding letter grades.
"""
def grade_classifier(score, custom_scale = None):
    """
    Grading Scales:
    - Standard Grading Scale:
        - score >= 90: 'A'
        - 80 <= score < 90: 'B'
        - 70 <= score < 80: 'C'
        - 60 <= score < 70: 'D'
        - score < 60: 'F'

    - Custom Grading Scale: Defined as a dictionary where the keys are score ranges and the values are corresponding letter grades.
        - key: (lower limit of score range, upper limit of score range)
        - value: letter grade
    """

    if custom_scale is None:
        # Standard grading scale
        if score >= 90:
            return 'A'
        elif 80 <= score < 90:
            return 'B'
        elif 70 <= score < 80:
            return 'C'
        elif 60 <= score < 70:
            return 'D'
        else:
            return 'F'
    else:
        # Custom grading scale
        for key in custom_scale:
            if key[0] <= score < key[1]:
                return custom_scale[key]
        return 'F'  # If the score doesn't fall into any custom scale range, return 'F' by default


# Demonstrate the custom grading scale
custom_grading_scale = {
    (95, 101): 'A+',
    (90, 95): 'A',
    (85, 90): 'A-',
    (80, 85): 'B+',
    (75, 80): 'B',
    (70, 75): 'B-',
    (65, 70): 'C+',
    (60, 65): 'C',
}

# Test the extended grade_classifier function with different scores using the custom grading scale
test_scores = [100, 95, 90, 89, 85, 77, 64, 55]
for score in test_scores:
    print(f"For a score of {score}, the corresponding grade is: {grade_classifier(score, custom_grading_scale)}")
```

```python
"""
Question 5:
You are given a list of integers, and your task is to implement a Python function that finds the longest subsequence of consecutive integers
within the list. A subsequence is defined as a sequence of integers in the same order as they appear in the original list. You should return
this longest consecutive subsequence as a list.

Parameters:
input_list (list): The list of integers.

Returns:
list: The longest consecutive subsequence.

Constraints:
- The input list will have at most a million integers.
- The integers in the list can be both positive and negative.
- The list can be unsorted, and the order of the subsequence should follow the original order in the list.
- If there are multiple subsequences of the same maximum length, the function returns the first one encountered.
"""

def find_longest_consecutive_subsequence(input_list):
    # define some variables
    max_length = 0
    result = []

    for i in range(0, len(input_list) - 1):
        current_length = 1
        current_subsequence = [input_list[i]]
    
        # check whether there is other consecutive element existing
        for j in range(i + 1, len(input_list)):
            if input_list[j] - current_subsequence[-1] == 1 or input_list[j] - current_subsequence[-1] == -1: 
                current_length += 1
                current_subsequence.append(input_list[j])

        # update max_length & result
        if current_length > max_length:
            max_length = current_length
            result = current_subsequence.copy()

    return result

# Test the function with the sample problem and additional test cases
input_list_1 = [1, 2, 3, 5, 6, 7, 8, 10, 11, 12]
result_1 = find_longest_consecutive_subsequence(input_list_1)
print(result_1)  # Output should be [5, 6, 7, 8]

input_list_2 = [1, 2, 3, 0, 5, 6, 7, 8]
result_2 = find_longest_consecutive_subsequence(input_list_2)
print(result_2)  # Output should be [5, 6, 7, 8]

input_list_3 = [1, 3, 5, 7, 9]
result_3 = find_longest_consecutive_subsequence(input_list_3)
print(result_3)  # Output should be [1]

input_list_4 = [5, 4, 2, 1, 0, 9, 8, 7, 6]
result_4 = find_longest_consecutive_subsequence(input_list_4)
print(result_4)  # Output should be [9, 8, 7, 6]

input_list_5 = [100, 3, 4, 200, 5, 6, 1, 3, 2]
result_5 = find_longest_consecutive_subsequence(input_list_5)
print(result_5)  # Output should be [3, 4, 5, 6]

input_list_6 = [-1, 3, 7, 2, 5, 8, 4, 6, 0, 1]
result_6 = find_longest_consecutive_subsequence(input_list_6)
print(result_6)  # Output should be [-1, 0, 1]
```
